### EDA - Exploratory Data Analysis
- Check its structure and integrity: type of values and missing values

```python
# Data structure:
display('Data Structure', pd.read_sql_query("select * from 'diamonds-train' limit 5;", conn))

# Type of Values:
display('Type of values', pd.read_sql_query("select typeof(carat), typeof(cut), typeof(clarity), typeof(x), typeof(y), typeof(z), typeof(depth), typeof('table'), typeof(price)  from 'diamonds-train' limit 1;", conn))
display(pd.read_sql_query("PRAGMA table_info('diamonds-train');", conn))

# Missing Values:
display('Missing Values', pd.read_sql_query("select * from 'diamonds-train' where carat is null or cut is null or color is null or clarity is null or x is null or y is null or z is null or depth is null or 'table' is null or price is null;", conn))
```

- Statistical summary of each column

```python
# Carat
display('Carat average', pd.read_sql_query("select avg(carat) from 'diamonds-train';", conn))
display('Carat variance', pd.read_sql_query("SELECT AVG(('diamonds-train'.carat - sub.a) * ('diamonds-train'.carat - sub.a)) as 'var(carat)' from 'diamonds-train', (SELECT AVG(carat) AS a FROM 'diamonds-train') AS sub;", conn))
display('Carat min', pd.read_sql_query("select min(carat) from 'diamonds-train';", conn))
display('Carat max', pd.read_sql_query("select max(carat) from 'diamonds-train';", conn))

# Cut
statement = "select count(*), cut from 'diamonds-train' group by cut;"
display(statement, pd.read_sql_query(statement, conn))

statement = "select count(*), cut, color, clarity from 'diamonds-train' group by cut, color, clarity;"
display(statement, pd.read_sql_query(statement, conn))

statement = "select count(cut) from 'diamonds-train';"
display(statement, pd.read_sql_query(statement, conn))

statement = "select count(distinct cut) from 'diamonds-train';"
display(statement, pd.read_sql_query(statement, conn))
#cursor = conn.cursor()
#cursor.execute("SELECT 'diamonds-test';")
#print(sqlite_stat1.stat)
#for line in cursor.fetchall():
#    print(line)
```
